```python
# importing packages
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Function to make the soup and get the page contents 
def soup_links(link):
    import requests
    requests = requests.get(link)
    src = requests.content
    soup = BeautifulSoup(src, 'lxml')
    return soup

# Function for getting a list of the desired values
def search(tag, attr):
    infos = soup.find_all(tag, attr)
    info = [info.text for info in infos]
    return info

# lists to add values in
jobs_title = []
Companies = []
addresses = []
dates = []
links = []

# Getting a job to search for
job = input('Enter a Job to search for on Wazzuf: ').lower().strip()
jobs_search = job.replace(' ', '%20')

# Looping through all the search results from the different pages
page_num = 0
while True:
    link = f'https://wuzzuf.net/search/jobs/?q={jobs_search}&a=navbg&start={page_num}'
    soup = soup_links(link)
    
    # Getting the desired values
    jobs_titles = search('a', {'class': 'css-nn640c', 'target': "_blank", 'rel': "noreferrer"})
    Companiess = search('a', {'class': 'css-17s97q8'})
    addressess = search('span', {'class': 'css-5wys0k'})

    datess = search('div', {'class': 'css-do6t5g'})
    dates_2 = search('div', {'class': 'css-4c4ojb'})
    datess.extend(dates_2)

    linkss = soup.find_all('a', {'class': 'css-nn640c', 'target': "_blank", 'rel': "noreferrer"})
    linkss = [link.get('href') for link in linkss]    
    
    # Extending lists with the output of each page
    jobs_title.extend(jobs_titles)
    Companies.extend(Companiess)
    addresses.extend(addressess)
    dates.extend(datess)
    links.extend(linkss)    
    
    # stopping the loop when reaching the limit
    page_num += 1
    tot_res = int(soup.find('strong').text.replace(',', ''))
    
    num_res_page = 15 # number of results per page
    pages_limit = tot_res // num_res_page # Getting the total number of pages then adding one for extra values
    
    if tot_res % num_res_page != 0:
        if page_num  <= pages_limit + 1:
            print(f'Getting Page {page_num} out of {pages_limit + 1} with a total of {tot_res} jobs')
        elif page_num  > pages_limit + 1:
            break
    else:
        if page_num  <= pages_limit:
            print(f'Getting Page {page_num} out of {pages_limit} with a total of {tot_res} jobs')
        elif page_num  > pages_limit :
            break

# Making a dataframe            
df = pd.DataFrame()
df['jobs_title'] = jobs_title 
df['Companies'] = Companies
df['addresses'] = addresses
df['dates'] = dates
df['links'] = links

# Making a Csv File
job_name = job.replace(' ', '_')
df.to_csv(f'Wazzuf_search_{job_name}.csv', index=False)
print('Finished')
```
